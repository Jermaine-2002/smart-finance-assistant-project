````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# This function loads transaction data from a CSV file and prepares it for analysis.

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant.

    Purpose:
        This function reads a CSV file and prepares it for financial analysis by:
        - checking required columns
        - cleaning currency values in the Amount column
        - handling missing values
        - converting Date and Amount into usable formats
        - removing invalid rows where necessary

    Required columns:
        Date, Amount, Category, Description

    Args:
        file_path: Path to a CSV file or uploaded file object

    Returns:
        pandas.DataFrame: Cleaned transaction data ready for analysis

    Raises:
        ValueError: If required columns are missing or no valid transaction data remains
    """

    import pandas as pd

    # -------------------------------------------------------------------------
    # STEP 1: Load the CSV file
    # -------------------------------------------------------------------------
    try:
        df = pd.read_csv(file_path)
        print("✅ CSV file loaded successfully.")
    except FileNotFoundError:
        raise ValueError("The transaction file could not be found. Please check the file path and try again.")
    except pd.errors.EmptyDataError:
        raise ValueError("The uploaded CSV file is empty. Please upload a file with transaction data.")
    except Exception as e:
        raise ValueError(f"Unable to read the CSV file. Please make sure it is a valid CSV. Error: {e}")

    # -------------------------------------------------------------------------
    # STEP 2: Check that all required columns exist
    # -------------------------------------------------------------------------
    required_columns = ['Date', 'Amount', 'Category', 'Description']
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        raise ValueError(
            f"The CSV file is missing required column(s): {missing_columns}. "
            f"Please make sure your file includes Date, Amount, Category, and Description."
        )

    print("✅ Required columns are present.")

    # -------------------------------------------------------------------------
    # STEP 3: Make a copy so the original data is not changed directly
    # -------------------------------------------------------------------------
    df = df.copy()

    # -------------------------------------------------------------------------
    # STEP 4: Clean the Amount column
    # -------------------------------------------------------------------------
    # Convert Amount to string first so string cleaning methods can be applied safely
    df['Amount'] = df['Amount'].astype(str).str.strip()

    # Remove dollar signs and commas
    df['Amount_Clean'] = (
        df['Amount']
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )

    # Convert cleaned amounts to numeric
    # Invalid values become NaN instead of causing a crash
    df['Amount_Clean'] = pd.to_numeric(df['Amount_Clean'], errors='coerce')

    print("✅ Amount column cleaned and converted to numeric format.")

    # -------------------------------------------------------------------------
    # STEP 5: Clean the Date column
    # -------------------------------------------------------------------------
    # Convert Date to datetime; invalid dates become NaT
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    print("✅ Date column converted to datetime format.")

    # -------------------------------------------------------------------------
    # STEP 6: Handle missing text values in Category and Description
    # -------------------------------------------------------------------------
    # Fill missing categories and descriptions with default labels
    df['Category'] = df['Category'].fillna('Uncategorised').astype(str).str.strip()
    df['Description'] = df['Description'].fillna('No description provided').astype(str).str.strip()

    # Replace blank strings with defaults as well
    df.loc[df['Category'] == '', 'Category'] = 'Uncategorised'
    df.loc[df['Description'] == '', 'Description'] = 'No description provided'

    print("✅ Missing Category and Description values handled.")

    # -------------------------------------------------------------------------
    # STEP 7: Remove rows with invalid or missing critical values
    # -------------------------------------------------------------------------
    initial_row_count = len(df)

    # Date and Amount are critical for analysis, so invalid rows are removed
    df = df.dropna(subset=['Date', 'Amount_Clean'])

    removed_rows = initial_row_count - len(df)

    if removed_rows > 0:
        print(f"⚠️ {removed_rows} row(s) were removed because they had invalid or missing Date/Amount values.")

    # -------------------------------------------------------------------------
    # STEP 8: Final validation
    # -------------------------------------------------------------------------
    if df.empty:
        raise ValueError(
            "No valid transaction data remains after cleaning. "
            "Please check that your CSV contains valid dates and amounts."
        )

    # -------------------------------------------------------------------------
    # STEP 9: Optional sorting for neat presentation
    # -------------------------------------------------------------------------
    df = df.sort_values(by='Date').reset_index(drop=True)

    print("✅ Transaction data cleaned successfully and is ready for analysis.")

    return df

# Save sample data to a CSV file for testing
df_sample.to_csv("sample_transactions.csv", index=False)

# Test the cleaning function
clean_data = load_and_clean_transaction_data("sample_transactions.csv")

print("\n📋 Cleaned Transaction Data:")
print(clean_data)
# Test your function
# clean_data = load_and_clean_transaction_data(df_sample)
# print(clean_data)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

# 🤖 AI Collaboration: Spending Analysis Function
# This function analyses cleaned transaction data and generates financial insights.

def analyze_spending_patterns(df):
    """
    Analyse spending patterns and generate business-style financial insights.

    Purpose:
        This function takes cleaned transaction data and calculates:
        - total spending by category
        - average transaction amount by category
        - transaction count by category
        - percentage of total spending by category
        - top spending categories
        - refund totals

    It also generates user-friendly insights based on the spending patterns.

    Args:
        df (pandas.DataFrame): Cleaned transaction data

    Returns:
        dict: A dictionary containing summaries, totals, top categories, and insights
    """

    import pandas as pd

    # -------------------------------------------------------------------------
    # STEP 1: Validate the input DataFrame
    # -------------------------------------------------------------------------
    if df is None or df.empty:
        raise ValueError("No transaction data is available for analysis. Please upload and clean a CSV file first.")

    required_columns = ['Date', 'Category', 'Description', 'Amount_Clean']
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        raise ValueError(
            f"The cleaned dataset is missing required column(s): {missing_columns}. "
            f"Please make sure the data cleaning step completed successfully."
        )

    print("✅ Cleaned transaction data is valid for analysis.")

    # -------------------------------------------------------------------------
    # STEP 2: Separate positive spending from refunds/negative transactions
    # -------------------------------------------------------------------------
    spending_df = df[df['Amount_Clean'] > 0].copy()
    refunds_df = df[df['Amount_Clean'] < 0].copy()

    if spending_df.empty:
        raise ValueError(
            "No positive spending transactions were found in the dataset. "
            "Please check whether the uploaded file contains valid expense data."
        )

    print("✅ Spending transactions separated from refunds.")

    # -------------------------------------------------------------------------
    # STEP 3: Calculate overall totals
    # -------------------------------------------------------------------------
    total_spending = spending_df['Amount_Clean'].sum()
    total_refunds = refunds_df['Amount_Clean'].sum() if not refunds_df.empty else 0
    transaction_count = len(spending_df)
    average_transaction = spending_df['Amount_Clean'].mean()

    print("✅ Overall spending totals calculated.")

    # -------------------------------------------------------------------------
    # STEP 4: Group spending by category
    # -------------------------------------------------------------------------
    category_summary = spending_df.groupby('Category').agg(
        Total_Spent=('Amount_Clean', 'sum'),
        Average_Amount=('Amount_Clean', 'mean'),
        Transaction_Count=('Amount_Clean', 'count')
    ).round(2)

    # -------------------------------------------------------------------------
    # STEP 5: Calculate percentage of total spending for each category
    # -------------------------------------------------------------------------
    category_summary['Percentage_of_Total'] = (
        (category_summary['Total_Spent'] / total_spending) * 100
    ).round(2)

    # Sort from highest to lowest spending
    category_summary = category_summary.sort_values(by='Total_Spent', ascending=False)

    print("✅ Spending grouped by category and percentages calculated.")

    # -------------------------------------------------------------------------
    # STEP 6: Identify top spending categories
    # -------------------------------------------------------------------------
    top_category = category_summary.index[0]
    top_category_total = category_summary.iloc[0]['Total_Spent']
    top_category_percentage = category_summary.iloc[0]['Percentage_of_Total']

    top_3_categories = category_summary.head(3)

    print("✅ Top spending categories identified.")

    # -------------------------------------------------------------------------
    # STEP 7: Generate financial insights
    # -------------------------------------------------------------------------
    insights = []

    insights.append(
        f"Total positive spending is ${total_spending:,.2f} across {transaction_count} transaction(s), "
        f"with an average transaction value of ${average_transaction:,.2f}."
    )

    insights.append(
        f"The highest spending category is '{top_category}', with total spending of "
        f"${top_category_total:,.2f}, representing {top_category_percentage:.2f}% of all positive spending."
    )

    # Identify categories with frequent small expenses
    frequent_categories = category_summary[
        (category_summary['Transaction_Count'] >= 3) &
        (category_summary['Average_Amount'] < average_transaction)
    ]

    for category in frequent_categories.index:
        insights.append(
            f"'{category}' shows repeated lower-value spending, which may indicate small purchases "
            f"that build up over time."
        )

    # Highlight very large spending categories
    high_percentage_categories = category_summary[category_summary['Percentage_of_Total'] >= 30]

    for category in high_percentage_categories.index:
        if category != top_category:
            insights.append(
                f"'{category}' also represents a significant share of spending at "
                f"{category_summary.loc[category, 'Percentage_of_Total']:.2f}% of the total."
            )

    # Refund insight
    if total_refunds < 0:
        insights.append(
            f"The dataset includes refunds or negative transactions totalling ${abs(total_refunds):,.2f}, "
            f"which reduce net outflows and should be considered separately from regular spending."
        )

    print("✅ Financial insights generated.")

    # -------------------------------------------------------------------------
    # STEP 8: Package results into a dictionary
    # -------------------------------------------------------------------------
    analysis_results = {
        'total_spending': round(total_spending, 2),
        'total_refunds': round(total_refunds, 2),
        'transaction_count': int(transaction_count),
        'average_transaction': round(average_transaction, 2),
        'category_summary': category_summary,
        'top_category': top_category,
        'top_category_total': round(top_category_total, 2),
        'top_category_percentage': round(top_category_percentage, 2),
        'top_3_categories': top_3_categories,
        'insights': insights
    }

    print("✅ Spending analysis complete.")

    return analysis_results

analysis = analyze_spending_patterns(clean_data)

print("\n📊 Category Summary:")
print(analysis['category_summary'])

print("\n💡 Financial Insights:")
for insight in analysis['insights']:
    print("-", insight)
# Test your function
# analysis = analyze_spending_patterns(clean_data)
# print(analysis)
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

# 🤖 AI Collaboration: Financial Recommendations Function
# This function reads analysis results and generates professional financial advice.

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis.

    Purpose:
        This function takes analysed spending data and creates:
        - observations about spending behaviour
        - possible savings opportunities
        - practical budgeting advice
        - user-friendly recommendations for a personal finance app

    Args:
        analysis_data (dict): Dictionary returned by analyze_spending_patterns()

    Returns:
        str: A formatted recommendations report
    """

    # -------------------------------------------------------------------------
    # STEP 1: Validate the analysis input
    # -------------------------------------------------------------------------
    if not analysis_data:
        raise ValueError(
            "No analysis data was provided. Please run the spending analysis step before generating recommendations."
        )

    required_keys = [
        'total_spending',
        'total_refunds',
        'transaction_count',
        'average_transaction',
        'category_summary',
        'top_category',
        'top_category_total',
        'top_category_percentage',
        'insights'
    ]

    missing_keys = [key for key in required_keys if key not in analysis_data]

    if missing_keys:
        raise ValueError(
            f"The analysis data is missing required information: {missing_keys}. "
            f"Please make sure the spending analysis step completed successfully."
        )

    print("✅ Analysis data validated successfully.")

    # -------------------------------------------------------------------------
    # STEP 2: Extract important values from the analysis results
    # -------------------------------------------------------------------------
    total_spending = analysis_data['total_spending']
    total_refunds = analysis_data['total_refunds']
    transaction_count = analysis_data['transaction_count']
    average_transaction = analysis_data['average_transaction']
    category_summary = analysis_data['category_summary']
    top_category = analysis_data['top_category']
    top_category_total = analysis_data['top_category_total']
    top_category_percentage = analysis_data['top_category_percentage']
    insights = analysis_data['insights']

    print("✅ Key spending metrics extracted from analysis data.")

    # -------------------------------------------------------------------------
    # STEP 3: Start building the recommendation report
    # -------------------------------------------------------------------------
    recommendations = []

    recommendations.append("SMART FINANCE ASSISTANT – PERSONALISED RECOMMENDATIONS")
    recommendations.append("=" * 65)
    recommendations.append("")
    recommendations.append("1. Spending Overview")
    recommendations.append(
        f"- Your total positive spending is ${total_spending:,.2f} across {transaction_count} transaction(s)."
    )
    recommendations.append(
        f"- Your average transaction amount is ${average_transaction:,.2f}."
    )

    if total_refunds < 0:
        recommendations.append(
            f"- Your records also include refunds totalling ${abs(total_refunds):,.2f}, which slightly reduce your net outflow."
        )

    recommendations.append("")

    print("✅ Spending overview section created.")

    # -------------------------------------------------------------------------
    # STEP 4: Add observations about top spending categories
    # -------------------------------------------------------------------------
    recommendations.append("2. Key Spending Observations")
    recommendations.append(
        f"- Your highest spending category is '{top_category}', with total spending of ${top_category_total:,.2f}."
    )
    recommendations.append(
        f"- This category represents {top_category_percentage:.2f}% of your total positive spending."
    )

    if top_category_percentage >= 40:
        recommendations.append(
            f"- This suggests that a large share of your spending is concentrated in one category. Reviewing '{top_category}' may create the biggest opportunity for savings."
        )
    elif top_category_percentage >= 25:
        recommendations.append(
            f"- '{top_category}' is a significant spending area. Even small reductions in this category could improve your overall cash flow."
        )
    else:
        recommendations.append(
            f"- Your spending appears to be more distributed across categories, which may indicate balanced spending habits."
        )

    recommendations.append("")

    print("✅ Key spending observations section created.")

    # -------------------------------------------------------------------------
    # STEP 5: Identify savings opportunities from category patterns
    # -------------------------------------------------------------------------
    recommendations.append("3. Savings Opportunities")

    high_spending_categories = category_summary[category_summary['Percentage_of_Total'] >= 20]
    frequent_categories = category_summary[category_summary['Transaction_Count'] >= 3]

    if not high_spending_categories.empty:
        recommendations.append(
            "- The following categories account for a notable share of your spending and may be worth reviewing:"
        )

        for category in high_spending_categories.index:
            total = category_summary.loc[category, 'Total_Spent']
            percentage = category_summary.loc[category, 'Percentage_of_Total']
            recommendations.append(
                f"  • {category}: ${total:,.2f} ({percentage:.2f}% of total spending)"
            )
    else:
        recommendations.append(
            "- No single category currently dominates your spending, which is a positive sign of balance."
        )

    recommendations.append("")

    if not frequent_categories.empty:
        recommendations.append(
            "- Some categories show repeated spending activity. Frequent smaller purchases can quietly add up over time:"
        )

        for category in frequent_categories.index:
            count = int(category_summary.loc[category, 'Transaction_Count'])
            avg = category_summary.loc[category, 'Average_Amount']
            recommendations.append(
                f"  • {category}: {count} transaction(s), average ${avg:,.2f} each"
            )
    else:
        recommendations.append(
            "- Your transaction pattern does not show many repeated small-purchase categories at this stage."
        )

    recommendations.append("")

    print("✅ Savings opportunities section created.")

    # -------------------------------------------------------------------------
    # STEP 6: Generate actionable financial advice
    # -------------------------------------------------------------------------
    recommendations.append("4. Actionable Advice")

    if top_category_percentage >= 40:
        recommendations.append(
            f"- Start by reviewing your spending in '{top_category}'. Since it makes up a large portion of your budget, this is likely your strongest savings opportunity."
        )
    else:
        recommendations.append(
            "- Continue monitoring your larger spending categories and look for areas where small adjustments can be made consistently."
        )

    if not frequent_categories.empty:
        recommendations.append(
            "- Consider setting a weekly or monthly limit for categories with repeated purchases."
        )
        recommendations.append(
            "- Reviewing frequent purchases can help identify habits that feel minor individually but become costly over time."
        )

    recommendations.append(
        "- Compare essential spending (such as groceries or transport) with discretionary spending (such as dining, entertainment, or coffee) to identify flexible savings opportunities."
    )
    recommendations.append(
        "- If you want stronger control over your cash flow, consider setting category-based budgets and checking your spending regularly."
    )

    if total_refunds < 0:
        recommendations.append(
            "- Because refunds are present in your transaction history, it may be helpful to track them separately so you can better distinguish true spending from money returned."
        )

    recommendations.append("")

    print("✅ Actionable advice section created.")

    # -------------------------------------------------------------------------
    # STEP 7: Include analysis insights in the final report
    # -------------------------------------------------------------------------
    recommendations.append("5. Additional Insights")
    for insight in insights:
        recommendations.append(f"- {insight}")

    recommendations.append("")

    print("✅ Additional insights section added.")

    # -------------------------------------------------------------------------
    # STEP 8: Add a professional closing summary
    # -------------------------------------------------------------------------
    recommendations.append("6. Final Recommendation Summary")
    recommendations.append(
        "Your spending analysis shows useful opportunities to improve financial awareness and cash flow control."
    )
    recommendations.append(
        "The most effective next step is to focus on high-impact categories first, then monitor repeated smaller purchases over time."
    )
    recommendations.append(
        "Consistent review of your transaction data can help you build stronger budgeting habits and make more confident financial decisions."
    )

    print("✅ Financial recommendations report completed.")

    # Join the list into one clean formatted string
    return "\n".join(recommendations)

recommendations_report = generate_financial_recommendations(analysis)

print(recommendations_report)
# Test your function
# recommendations = generate_financial_recommendations(analysis)
# print(recommendations)
````
